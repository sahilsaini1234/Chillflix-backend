```python
import ast
def genere(object):
    list=[]
    for i in ast.literal_eval(object):
        list.append(i['name'])
    return list
data['genres']=data['genres'].apply(genere)
```

```python
def extract(object):
    counter=0
    l=[]
    for i in ast.literal_eval(object):
        if counter < 4:
            l.append(i['name'])
        else:
            break
        counter+=1
    return l   
```

```python
data['cast']=data['cast'].apply(extract)
```

```python
def extract_crew(object):
    l=[]
    for i in ast.literal_eval(object):
        if(i['job']=='Director'):
            l.append(i['name'])
    for i in ast.literal_eval(object):
        if(i['job']=='Writer'):
            l.append(i['name'])
    return l        
            
```

```python
data['crew']=data['crew'].apply(extract_crew)
```

```python
def space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
data['crew']=data['crew'].apply(space)
data['genres']=data['genres'].apply(space)
data['cast']=data['cast'].apply(space)
data['keywords']=data['keywords'].apply(space)

```
